```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', color_codes=True)

np.random.seed(sum(map(ord, 'categorical')))
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)# 偏移
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.swarmplot(x='total_bill', y='day', hue='time', data=tips)
```

```python
# 盒图
# IQR即统计学概念四分位距，第一/四分位与第三/四分位之间的距离
# N = 1.5IQR，如果一个值>Q3+N或<Q1-N，则为离群点
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
sns.violinplot(x='total_bill', y='day', hue='time', data=tips)# 小提琴图
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, innner=None)
sns.swarmplot(x='day', y='total_bill', data=tips, color='w', alpha=.5)# 两个图结合在一起
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.pointplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.pointplot(x='class', y='survived', hue='sex', data=titanic,
              palette={'male':'g', 'female':'m'},
              markers=['^', 'o'], linestyles=['-', '--'])
```

```python
sns.boxplot(data=iris, orient='h')# 加上orient横过来
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', hue='smoker',
               col='time', data=tips, kind='swarm')
```

```python
sns.factorplot(x='time', y='total_bill', hue='smoker',
               col='day', data=tips, kind='box', size=4, aspect=.5)# 大小，长宽比
```
