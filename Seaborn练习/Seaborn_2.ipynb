```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats, integrate

sns.set(color_codes=True)
np.random.seed(sum(map(ord, 'distributions')))# 随机生成高斯数据
```

```python
x = np.random.normal(size=100)
sns.distplot(x, kde=False)
```

```python
sns.distplot(x, bins=20, kde=False)
```

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma)
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=['x', 'y'])
df
```

```python
sns.jointplot(x='x', y='y', data=df)# 绘制散点图
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style('white'):
    sns.jointplot(x=x, y=y, kind='hex', color='k')
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)# 对角线是单变量情况，非对角线为两个数据之间情况
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)# 绘制回归关系
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.regplot(data=tips, x='size', y='tip')
```

```python
sns.regplot(data=tips, x='size', y='tip', x_jitter=.05)# 在原始点上加上小范围浮动
```
