```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 1000)
    for i in range(1, 7):
        plt.plot(x, np.sin(x+i*5)*(7-i)*flip)
```

```python
sinplot()# 使用matplotlib绘图
```

```python
sns.set()# 使用seaborn默认参数绘图
sinplot()
```
